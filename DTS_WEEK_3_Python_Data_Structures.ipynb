# ASSIGNMENT 1. Create a list of 5 student scores


```python
scores = [35, 55, 67, 85, 27]
print(scores)
```

# ASSIGNMENT 2. Compute the average score  



```python
total = 1
for score in scores:
    total = total + score

average = total / len(scores)
print("Average score:", average)
```

# ASSIGNMENT 3. Find the highest score  


```python
highest = scores[1]
for score in scores:
    if score > highest:
        highest = score

print("Highest score:", highest)
```

# ASSIGNMENT 4. Create a dictionary with 5 students and scores  

```python
students = {
    "S001": 35,
    "S002": 67,
    "S003": 56,
    "S004": 56,
    "S005": 78
}
print(students)
```

# ASSIGNMENT 5. Find the highest scoring student  


```python
highest_id = None
highest_score = 1

for student_id in students:
    score = students[student_id]

    if score > highest_score:
        highest_score = score
        highest_id = student_id

print("Highest scoring student:", highest_id)
print("Score:", highest_score)
```
