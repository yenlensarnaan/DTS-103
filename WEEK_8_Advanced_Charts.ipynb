```python

import matplotlib.pyplot as plt
import numpy as np
```

```python

students = ["Amos", "Henry", "Joshua", "Rex", "Raymond", "Peace", "Confidence", "Michael", "Misheal", "Anna", "Abigail", "Betty", "Samuel", "David", "John"]
chemistry_scores = [50, 70, 60, 56, 45, 67, 80, 90, 81, 46, 78, 39, 87, 77, 59]
history_scores = [71, 65, 40, 58, 70, 38, 91, 66, 85, 59, 39, 29, 88, 77, 61]
biology_scores = [69, 48, 38, 72, 59, 65, 55, 90, 94, 77, 40, 59, 44, 50, 49,]
```

```python

plt.figure()
plt.hist(chemistry_scores, bins=5)
plt.title("Distribution of Chemistry Scores")
plt.xlabel("Chemistry Scores")
plt.ylabel("Number of students")
plt.show()
```

```python

plt.figure()
plt.scatter(chemistry_scores, history_scores)
plt.title("Chemistry vs History Scores")
plt.xlabel("Chemistry Scores")
plt.ylabel("History Scores")
plt.show()
```

```python

plt.figure()
plt.plot(students, chemistry_scores)
plt.plot(students, history_scores)
plt.plot(students, biology_scores)
plt.title("Chemistry, History and biology Scores Comparison")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python

plt.figure()

plt.subplot(1, 2, 1)
plt.hist(biology_scores, bins=5)
plt.title("Biology Histogram")

plt.subplot(1, 2, 2)
plt.scatter(chemistry_scores, biology_scores)
plt.title("Chemistry vs Biology")

plt.tight_layout()
plt.show()
```

Histogram shows that most students scored between 50 and 58 in chemistry
Scatter plot indicates that students score higher in chemistry
Multiple line chart compare the  performance of students in chemistry history and biology it shows that many students perform best in chemistry
The subplot provide a clear comparison of biology score distribution and the relationship between biology and chemistry
