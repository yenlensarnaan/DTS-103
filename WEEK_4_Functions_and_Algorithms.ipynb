```python

def compute_average(scores):
    total = 0

    for score in scores:
        total = total + score

    average = total / len(scores)

    return average


scores = [25, 55, 75, 65, 90, 45]

avg = compute_average(scores)

print("Average score:", avg)
```

```python

def check_pass_fail(score):

    if score >= 50:
        return "Pass"
    else:
        return "Fail"


print(check_pass_fail(90))
print(check_pass_fail(50))
```

```python

scores = [59, 39, 50, 70, 20]

for score in scores:
    result = check_pass_fail(score)
    print(score, result)
```
