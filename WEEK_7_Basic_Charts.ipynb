```python
import matplotlib.pyplot as plt
```

```python

students =["Henry", "Joy", "Ezekiel", "Obadiah", "Luke", "Mark", "Job", "Raphael", "Hope", "Zara"]
french_scores = [59, 70, 85, 69, 67, 60, 90, 55, 44, 88]
geography_scores = [80, 50, 70, 49, 60, 39, 90, 74, 66, 80]
economics_scores = [60, 70, 80, 90, 50, 45, 77, 81, 49, 69]
```

```python

plt.figure()
plt.plot(students, french_scores)
plt.title("French Scores - Line Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python

plt.figure()
plt.bar(students, geography_scores)
plt.title("Geography Scores - Bar Chart")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```
