```python

file = open("week5_student_scores.csv", "r")

next(file)

total = 0
count = 0

for line in file:
    data = line.strip().split(",")
    math_score = int(data[2])

    total += math_score
    count += 1

average = total / count

print("Average Maths Score:", average)

file.close()
```
